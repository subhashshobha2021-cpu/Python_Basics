```python
# 6. Write a function to demonstrate the behavior of mutable and immutable arguments.
def immutable_data(data_type):
  if type(data_type) == list or type(data_type) == dict or type(data_type) == set:
    print("Data is mutable")
  else:
    print("Data is not mutable")

list_1 = [3,5,1,"subhash",(3,4,6),2.5,["deepak",3.4,5]]
tuple_1 = (3,5,1,"subhash",(3,4,6),2.5,["deepak",3.4,5])

immutable_data(list_1)
immutable_data(tuple_1)
```

```python
# 7. Write a program that performs basic arithmetic operations on two user-input numbers.

def arithmetic_ops(a,b):
  add_ = a+b
  sub_ = a-b
  mul_ = a*b
  div_ = a/b

  return add_, sub_, mul_, div_


arithmetic_ops(100,34)
```

```python
#. Write a Python program to find the sum of all even numbers between 1 and 50
total_summary = 0
for number in range(1,51):
  if number % 2 == 0:
    total_summary += number

print(total_summary)
```

```python
# Write a program to reverse a string using a while loop.
string_1 = "subhash is 31 years old"
index = len(string_1) - 1
reverse_string = ""
while index >= 0:
  reverse_string += string_1[index]
  index = index - 1
print(reverse_string)
```

```python
# . Write a Python program to calculate the factorial of a number provided by the user using a while loop.

factorial = 1
n = 4
while n > 0:
  factorial = factorial * n
  n = n - 1
print(factorial)
```
